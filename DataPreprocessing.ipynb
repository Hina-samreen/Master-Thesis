```python
import pandas as pd
```

### Reading raw data before processing

```python
df = pd.read_csv('GermanSet_interpretability.csv')
pd.options.display.max_colwidth = 500
df.head(15)
```

###  The below function cleans/processes the data such only the spans are added as negative/positive tokens

```python
def retreive_spans_with_param(param_for_spans):
    span_list = []
    import ast
    for i in df.index:
        if len(ast.literal_eval(df[param_for_spans][i])) == 0:
            span_list.append('...')
        if len(ast.literal_eval(df[param_for_spans][i])) < 2:
            for dict_ in (x for x in ast.literal_eval(df[param_for_spans][i])):
                if len(dict_['span'].split())>1:
                    if len(dict_.get('span').split('>')) > 1 :
                        span_list.append(dict_.get('span').split('>')[-1])
                    else:
                        span_list.append(dict_.get('span'))
                else:
                    span_list.append(dict_.get('span'))
        else:
            span_list_multi = []
            for dict_ in (x for x in ast.literal_eval(df[param_for_spans][i])):
                if len(dict_['span'].split())>1:
                    if len(dict_.get('span').split('>')) > 1 :
                        span_list_multi.append(dict_.get('span').split('>')[-1])
                    else:
                        span_list_multi.append(dict_.get('span'))
                else:
                    span_list_multi.append(dict_.get('span'))
            span_list.append(span_list_multi)
    #print(len(span_list))
    return span_list
    
```

### Applying the pre-processing to the below columns using the function "retreive_spans_with_param(param)", so that I can have only the spans and not the enire start and end strings, and assigning the results to new columns.

```python

df['negative object_spans'] = retreive_spans_with_param('span_annotation:::negative object')
df['negative spans'] = retreive_spans_with_param('span_annotation:::negative')
df['positive spans'] = retreive_spans_with_param('span_annotation:::positive')
df['positive object_spans'] = retreive_spans_with_param('span_annotation:::postive object')
```

### Dropping the columns "span_annotation:::negative object, span_annotation:::positive,	span_annotation:::negative,	span_annotation:::postive object" and saving to a new CSV. The pre-processing is now complete.

```python
df.drop(['span_annotation:::negative object', 'span_annotation:::positive'	,'span_annotation:::negative'	, 'span_annotation:::postive object'], inplace = True , axis = 1)
df.to_csv('PreProcessedDataGerman.csv', index = False)
```
